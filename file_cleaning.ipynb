```python
import re

with open("data/moliere_3.txt") as file:
    moliere = file.read()
```

```python
moliere = re.sub(r'( )+',' ',moliere)
moliere = re.sub(r'\[\d*\]','',moliere)
```

```python
example = moliere.split('\n')[315:-480]
example = [line.lstrip(' ') for line in example if line != '' and not line.lstrip(' ')[:2].isupper()]
example
```

```python
text = ""
for line in example:
    text += line + ' '
```

```python
with open('data/moliere_3_clean.txt','w') as file:
    file.write(text)
```
